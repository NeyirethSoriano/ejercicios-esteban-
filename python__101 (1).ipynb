#Creamos 3 ejemplos de diccionarios:

```python
numeros = (1, 2, 3, 4, 2, 5, 6)
frutas = ("manzana", "madarina", "uva", "manzana", "naranja", "manzana")
booleanos = (True, False, True, False, True)
print(numeros)
print(frutas)
print(booleanos)
```

##count():
###`Devuelve el número de veces que aparece un valor especificado en una tupla`

```python
print(numeros.count(2))
print(frutas.count("manzana"))
print(booleanos.count(True))
```

##index():
###`Busca en la tupla un valor específico y devuelve la posición donde se encontró.`

```python
print(numeros.index(3))
print(frutas.index("uva"))
print(booleanos.index(False))
```
